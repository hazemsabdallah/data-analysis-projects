# Project: Wrangle and Analyze Data @WeRateDogs
***
@WeRateDogs is a Twitter account popular for rating dogs, usually >10 out of 10, and adding a joke with the name and stage depending on the dog in the photo. Some insights about the account and its tweet contents have been collected, extracted, and generated using Twitters data  archive, Twitters API, and a dog breed classifying neural network. The below analyses are covering tweet data from November 2015 to July 2017.

To kick it off, @WeRateDogs has an average retweet count of 2,397 with a maximum of 75,549. Tweets made by the account has an average favorite count of 7,993 with a maximum of 152,631. For tweets with uniquely specified dog stage (i.e doggo, floofer, pupper, puppo), the below three bar charts show a comaprison between the 4 stages in terms of ratings, reweets, and favorites. Even though the average ratings for all stages are quite close, it looks like puppos is the highest dog stage in all 3 metrics. This is specially evident in average retweets and average favorits. On the other hand, how sad is it to see puppers always on the losing side?
